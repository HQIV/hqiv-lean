```python
import pandas as pd

import matplotlib.pyplot as plt

df = pd.read_csv('selected_benchmarks.csv')

for col in ['result', 'author', 'family']:
    counts = df[col].value_counts()
    misc_count = (counts == 1).sum()
    counts = counts[counts > 1]
    counts['misc'] = misc_count
    
    plt.figure(figsize=(10, 6))
    counts.plot(kind='bar')
    plt.title(f'Occurrences by {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
